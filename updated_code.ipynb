```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import math
from PIL import Image
path = 'images/sample3.jpg'
#path = 'images/sample2.webp'
#path = 'images/sample.jpg'
plt.rcParams["figure.figsize"] = (10,10)
```

```python
img = cv2.imread(path)
gr = cv2.imread(path, 0)
```

```python
plt.imshow(gr, cmap = 'gray')
plt.show()
```

```python
r, th = cv2.threshold(gr,50,255,cv2.THRESH_BINARY_INV)
plt.imshow(th, cmap = 'gray')
plt.show()
```

```python
cv2.imwrite('images/output.png', th)
```

```python
plt.imshow(img)
plt.show()
```

```python
mask = np.where(th == 255)
```

```python
RED = [255,0,0]
GREEN = [0,255,0]
BLUE = [0,0, 255]
```

```python
img[mask] = GREEN
plt.imshow(img)
plt.show()
```

```python
img[mask] = BLUE
plt.imshow(img)
plt.show()
```

```python
img[mask] = RED
plt.imshow(img)
plt.show()
```
