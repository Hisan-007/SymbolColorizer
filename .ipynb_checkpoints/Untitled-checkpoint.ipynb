```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.feature import hog
from skimage.transform import resize
from skimage.metrics import structural_similarity as ssim
from skimage import exposure
import random
plt.rcParams["figure.figsize"] = (10,10)
path = 'images/sample9.jpg'
```

```python
def gray(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def sharpen(img):
    kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])

    return cv2.filter2D(src=img, ddepth=-1, kernel=kernel)

def dilate(img):
    
    kernel = np.ones((3,3), np.uint8)
    
    return cv2.erode(img, kernel, iterations=1)
```

```python
img = cv2.imread(path)
timg = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img = cv2.imread(path, 0)

th = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,35,1)

sharp = sharpen(th)
ex = dilate(th)
#plt.imshow(timg[:,:,2])
#plt.figure()
plt.imshow(th, cmap = 'gray')
plt.figure()
plt.imshow(sharp, cmap = 'gray')
plt.figure()
plt.imshow(ex, cmap = 'gray')
```

```python
def dist(x,y):   
    return np.sqrt(np.sum((x-y)**2))

def get_feature(c):
    #c = contours[320]
    x, y, w, h = cv2.boundingRect(c)
    part = gray[y:y+h,x:x+w]
    part = cv2.resize(part, (50,50), interpolation = cv2.INTER_AREA)

    r, th = cv2.threshold(part,100,255,cv2.THRESH_BINARY_INV)
    fd, hog_image = hog(th, orientations=5, pixels_per_cell=(20, 20),
                    cells_per_block=(2, 2), visualize=True, multichannel=False)
    
    #plt.imshow(th, cmap = 'gray')
    #plt.figure()
    #plt.imshow(part, cmap = 'gray')
    
    h = has_symbol(th)
    #print(h)
    return fd#, hog_image

def white_count(img):
    mask = img > 0
    return mask.sum()

def has_symbol(img):
    w = white_count(img)
    if(w > 30):
        return True
    else:
        return False
    
def get_img(c):
    x, y, w, h = cv2.boundingRect(c)
    part = gray[y:y+h,x:x+w]
    part = cv2.resize(part, (50,50), interpolation = cv2.INTER_AREA)

    r, th = cv2.threshold(part,50,255,cv2.THRESH_BINARY_INV)
    orig = img[y:y+h,x:x+w]
    #orig = cv2.resize(orig, (50,50), interpolation = cv2.INTER_AREA)
    return orig, th

def disp2(i, j):
    img1 = get_img(contours[i])[0]
    img2 = get_img(contours[j])[0]
    plt.imshow(img1, cmap = 'gray')
    plt.figure()
    plt.imshow(img2, cmap = 'gray')
    
def is_square(img):
    #print(img.shape)
    h, w, d = img.shape
    if h//2 > w or w//2 > h:
        return False
    return True
```

```python
ind = []
d = {}
for i in range(len(contours)):
    #print(i)
    #print()
    flag = False
    for key in d:
        if i in d[key]:
            #print('key', i)
            flag = True
            break
    if flag:
        continue
    if(not is_square(get_img(contours[i])[0])):
        continue
    for j in range(i, len(contours)):
        
        if(not is_square(get_img(contours[j])[0])):
            continue
        if i == j:
            continue
        c1 = contours[i]
        c2 = contours[j]
        
        
        if(has_symbol(get_img(c1)[1]) and has_symbol(get_img(c2)[1])):
            #print(i,j)
            a = get_img(c1)[0]
            b = get_img(c2)[0]
            
            h1 = cv2.calcHist([a], [0, 1, 2], None, [8, 8, 8],
                    [0, 256, 0, 256, 0, 256])
            h2 = cv2.calcHist([b], [0, 1, 2], None, [8, 8, 8],
                    [0, 256, 0, 256, 0, 256])
            diff = cv2.compareHist(h1, h2, cv2.HISTCMP_BHATTACHARYYA)
            if(diff < 0.5):
                for key in d:
                    if i in d[key] and j not in d[key]:
                        #d[key].append(i)
                        d[key].append(j)
                        break
                else:
                    d[str(len(d))] = [i,j]
                ind.append((i,j))
```
